```python
#W6A1
a = list(map(int, input().split()))
b = set(a)
c = []
for x in a:
    if x in b:
        c.append(x)
        b.remove(x)
print(c)
```

```python
#W6A4 
s = input().split()
d = {} 
for i in range(len(s)):
    x, y = s[i].split(':')
    if x not in d:
        d[x] = []
    d[x].append(y)
print(d)
```

```python
#W6A9
n = int(input())
a = {}
for _ in range(n):
    k, v = input().split()
    a[k] = int(v)

m = int(input())
b = {}
for _ in range(m):
    k, v = input().split()
    b[k] = int(v)

c = {}

for k, v in a.items():
    c[k] = c.get(k, 0) + v
for k, v in b.items():
    c[k] = c.get(k, 0) + v

c = dict(sorted(c.items()))
print(c)
```

```python
#W6A10
a = list(map(int, input().split()))
k = int(input())
res = []

for i in range(len(a)):
    for j in range(i + 1, len(a)):
        if a[i] + a[j] == k:
            res.append((a[i], a[j]))

res.sort(key=lambda x: x[0])
print(res)
```

```python
#W6A11
a = tuple(map(int, input().split()))
even = []
odd = []

for x in a:
    if x % 2 == 0:
        even.append(x)
    else:
        odd.append(x)

print(tuple(even))
print(tuple(odd))
```

```python
#W6A12
a = list(map(int, input().split()))
d = {}

for x in a:
    d[x] = d.get(x, 0) + 1

m = max(d.values())
res = min(k for k, v in d.items() if v == m)
print(res)
```

```python
#W6A15
n = int(input())
a = {}

for _ in range(n):
    k, v = input().split()
    a[k] = int(v)

k = int(input())
b = {}

for x, y in a.items():
    if y > k:
        b[x] = y

print(b)
```
