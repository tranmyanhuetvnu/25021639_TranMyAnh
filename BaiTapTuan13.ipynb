```python
#W11A1
from pathlib import Path

WORKDIR = Path('.')
path = 'tma.txt'

def reverse(path: str) -> int:
    docfile = WORKDIR / path
    with open(docfile, 'r') as f:
        n = int(f.read().strip())
    n_rev = 0
    while n > 0:
        n_rev = n_rev * 10 + n % 10
        n //= 10
    return n_rev

print(reverse(path))
```

```python
#W11A4
from pathlib import Path

WORKDIR = Path('.')
path1 = 'a.txt'
path2 = 'b.txt'

def areAnagrams(path1: str, path2: str) -> bool:
    docfile1 = WORKDIR / path1
    docfile2 = WORKDIR / path2

    with open(docfile1, 'r', encoding='utf-8') as f:
        s1 = f.read().lower()
    with open(docfile2, 'r', encoding='utf-8') as f:
        s2 = f.read().lower()

    cnt1 = [0] * 26
    cnt2 = [0] * 26

    for c in s1:
        if 'a' <= c <= 'z':
            cnt1[ord(c) - ord('a')] += 1

    for c in s2:
        if 'a' <= c <= 'z':
            cnt2[ord(c) - ord('a')] += 1

    return cnt1 == cnt2

print(areAnagrams(path1, path2))
```

```python

def getMoney(path: str) -> list[int]:
    with open(path, 'r', encoding='utf-8') as f:
        return list(map(int, f.read().strip().split()))

def rob(houses: list[int]) -> int:
    if not houses:
        return 0
    if len(houses) == 1:
        return houses[0]
    prev2 = houses[0]
    prev1 = max(houses[0], houses[1])
    for i in range(2, len(houses)):
        cur = max(prev1, prev2 + houses[i])
        prev2 = prev1
        prev1 = cur
    return prev1
```

```python
def findMinMax(path: str):
    try:
        with open(path, 'r', encoding='utf-8') as f:
            data = list(map(int, f.read().split()))
        if not data:
            return
        print(min(data), max(data))
    except:
        print("Mission failed")
```

```python
def maximumProduct(path: str) -> int:
    with open(path, 'r', encoding='utf-8') as f:
        lines = f.read().split()
    n = int(lines[0])
    a = list(map(int, lines[1:]))

    a.sort()
    return max(a[-1] * a[-2] * a[-3], a[0] * a[1] * a[-1])
```

```python
def moveZeroes(fileName: str) -> list:
    with open(fileName, 'r', encoding='utf-8') as f:
        data = list(map(int, f.read().split()))
    n = data[0]
    a = data[1:1 + n]

    res = []
    zero_cnt = 0
    for x in a:
        if x == 0:
            zero_cnt += 1
        else:
            res.append(x)
    res.extend([0] * zero_cnt)
    return res
```

```python
def averageTime(path: str):
    total = 0.0
    cnt = 0
    with open(path, 'r', encoding='utf-8') as f:
        lines = f.readlines()[1:]
    for line in lines:
        parts = line.strip().split(',')
        total += float(parts[1])
        cnt += 1
    if cnt == 0:
        return 0.00
    return round(total / cnt, 2)
```

```python
def findMovies(path: str) -> list[str]:
    res = []
    with open(path, 'r', encoding='utf-8') as f:
        lines = f.readlines()[1:]
    for line in lines:
        parts = [x.strip() for x in line.strip().split(',')]
        movie = parts[1]
        description = parts[2]
        rating = float(parts[3])
        if rating > 8.0 and description.lower() != 'boring':
            res.append(movie)
    if not res:
        return ["All are boring!"]
    return res
```

```python
def printPerimeters(path: str):
    with open(path, 'r', encoding='utf-8') as f:
        lines = [line.strip() for line in f if line.strip()]

    i = 0
    while i < len(lines):
        shape = lines[i].split(':')[1].strip()
        info = lines[i + 1]
        if shape == 'SQUARE':
            side = float(info.split(':')[1])
            peri = 4 * side
        elif shape == 'RECTANGLE':
            parts = info.split()
            width = float(parts[1])
            height = float(parts[3])
            peri = 2 * (width + height)
        elif shape == 'CIRCLE':
            radius = float(info.split(':')[1])
            peri = 2 * 3.141592653589793 * radius
        else:
            peri = 0
        print(f"{peri:.2f}")
        i += 2
```

```python
def productExceptSelf(path: str) -> list[int]:
    with open(path, 'r', encoding='utf-8') as f:
        a = list(map(int, f.read().split()))
    n = len(a)
    res = [1] * n

    prefix = 1
    for i in range(n):
        res[i] = prefix
        prefix *= a[i]

    suffix = 1
    for i in range(n - 1, -1, -1):
        res[i] *= suffix
        suffix *= a[i]

    return res
```

```python
import math

def estimatedTime(path: str):
    with open(path, 'r', encoding='utf-8') as f:
        lines = [line.strip() for line in f if line.strip()]

    src = lines[0].split()
    des = lines[1].split()
    vel = lines[2].split()

    x1 = float(src[2])
    y1 = float(src[4])
    x2 = float(des[2])
    y2 = float(des[4])
    v = float(vel[1])

    distance = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
    time = distance / v
    return round(time, 2)
```

```python
def imageSmoother(path: str):
    with open(path, 'r', encoding='utf-8') as f:
        data = list(map(int, f.read().split()))
    m = data[0]
    a = []
    idx = 1
    for _ in range(m):
        a.append(data[idx:idx + m])
        idx += m

    res = [[0] * m for _ in range(m)]
    for i in range(m):
        for j in range(m):
            s = 0
            cnt = 0
            for dx in (-1, 0, 1):
                for dy in (-1, 0, 1):
                    ni = i + dx
                    nj = j + dy
                    if 0 <= ni < m and 0 <= nj < m:
                        s += a[ni][nj]
                        cnt += 1
            res[i][j] = s // cnt

    for i in range(m):
        print(' '.join(map(str, res[i])))
```
