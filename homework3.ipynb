```python
#bai4
h = int(input())
h -= 1
for i in range(0, h+1):
    for j in range(0, 2*h + 1):
        if j == h - i or j == h + i or i == h:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

```python
#bai6
import math
n = int(input())
for i in range(1, math.isqrt(n)+1):
    if n % i == 0:
        print(i, end = ' ')
        if i != n // i: 
            print(n//i, end = ' ')
```

```python
#bai10
import math
n = int(input())
tong = 0
for i in range(1, math.isqrt(n)+1):
    if n % i == 0:
        tong += i
        if i != n //i:
            tong += n //i 
if tong - n == n: 
    print('yes')
else: 
    print('no')
```
