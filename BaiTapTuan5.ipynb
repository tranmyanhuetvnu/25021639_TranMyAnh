```python
#W5A3
from math import *
def nt(a):
    for i in range(2, isqrt(a)+1):
        if a % i == 0:
            return False
    return True 
if __name__ == '__main__':
    a = int(input())
    print(nt(a))
```

```python
#W5A4
from math import *
def hh(n):
    tong = 0
    for i in range(1, isqrt(n)+1):
        if n % i == 0:
            tong += i
            if i != n//i:
                tong += n//i
    if tong - n == n:
        return True
    else:
        return False
if __name__ == '__main__':
    n= int(input())
    print(hh(n))
```

```python
#W5A6
def gt(n):
    res = 1
    for i in range(1, n+1):
        res *= i
    return res
n = int(input())
print(gt(n))
```

```python
#W5A7
def mt(a, b, c):
    if c == '+':
        return a + b
    elif c == '-':
        return a - b
    elif c == '*':
        return a * b
    elif c == '/':
        return a / b
a, b = map(int, input().split())
c = input()
print(mt(a,b,c))
```

```python
#W5A8
def kc(x, y):
    return bin(x^y).count('1')
x, y = map(int, input().split())
print(kc(x,y))
```

```python
#W5A9
def tcs(n):
    tong = 0
    while n != 0:
        r = n % 10
        tong += r
        n //= 10
    return tong
n = int(input())
print(tcs(n))
```
