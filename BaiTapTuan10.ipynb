```python
#W9A3
def min_by_one_swap(n):
    s = list(str(n))
    length = len(s)

    for i in range(length):
        mn = s[i]
        pos = -1

        for j in range(length - 1, i, -1):
            if s[j] < mn:
                if i == 0 and s[j] == '0':
                    continue
                mn = s[j]
                pos = j

        if pos != -1:
            s[i], s[pos] = s[pos], s[i]
            return int(''.join(s))

    s[-1], s[-2] = s[-2], s[-1]
    return int(''.join(s))
n = int(input())
print(min_by_one_swap(n))
```

```python
#W9A4
def can_transform(a, b):
    n = len(a)
    m = len(b)

    dp = [[False] * (m + 1) for _ in range(n + 1)]
    dp[0][0] = True

    for i in range(1, n + 1):
        if a[i - 1].islower():
            dp[i][0] = dp[i - 1][0]

    for i in range(1, n + 1):
        for j in range(0, m + 1):
            if a[i - 1].islower():
                dp[i][j] |= dp[i - 1][j]
                if j > 0 and a[i - 1].upper() == b[j - 1]:
                    dp[i][j] |= dp[i - 1][j - 1]
            else:
                if j > 0 and a[i - 1] == b[j - 1]:
                    dp[i][j] = dp[i - 1][j - 1]

    return dp[n][m]
Q = int(input())
for _ in range(Q):
    a = input().strip()
    b = input().strip()
    print("YES" if can_transform(a, b) else "NO")
```

```python
#W9A6
def next_permutation(a):
    n = len(a)
    i = n - 2
    while i >= 0 and a[i] >= a[i + 1]:
        i -= 1

    if i < 0:
        a.reverse()
        return

    j = n - 1
    while a[j] <= a[i]:
        j -= 1

    a[i], a[j] = a[j], a[i]
    a[i + 1:] = reversed(a[i + 1:])
n = int(input())
a = list(map(int, input().split()))
next_permutation(a)
print(*a)
```

```python
#W9A7
def canPlaceFlowers(flowerbed, k):
    n = len(flowerbed)

    for i in range(n):
        if flowerbed[i] == 0:
            left = (i == 0 or flowerbed[i-1] == 0)
            right = (i == n-1 or flowerbed[i+1] == 0)

            if left and right:
                flowerbed[i] = 1
                k -= 1
                if k == 0:
                    return True

    return k <= 0
flowerbed = list(map(int, input().split())) 
k = int(input())
print(canPlaceFlowers(flowerbed, k))
```

```python
#W9A10
import bisect

def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()
    res = 0

    for h in houses:
        i = bisect.bisect_left(heaters, h)

        dist1 = abs(h - heaters[i]) if i < len(heaters) else float('inf')
        dist2 = abs(h - heaters[i - 1]) if i > 0 else float('inf')

        res = max(res, min(dist1, dist2))

    return res
a = list(map(int, input().split()))
b = list(map(int, input().split()))
print(findRadius(a,b))
```
