```python
#W4A6
gh = 10 **6+1
p = [True] * (gh)
def nt():
    p[0], p[1] = False, False
    for i in range(2, int(gh**0.5) + 1):
        if p[i]:
            for j in range(i*i, 10**6 + 1, i):
                p[j] = False
F = [0] * (gh)
def ktao():
    F[2] = 2
    tong = 2 
    for i in range(3, gh):
        if p[i]:
            tong += i 
            F[i] = tong
        F[i] = tong
if __name__ == '__main__':
    nt()
    ktao()
    t = int(input())
    while t != 0:
        l, r = map(int, input().split())
        print(F[r] - F[l])
        t -= 1
```

```python
#W4A8
import math
def dn(n):
    tmp = 0
    while n != 0:
        r = n % 10
        tmp = tmp* 10 + r
        n //= 10
    return tmp
def dx(s):
    l = 0
    r = len(s) - 1
    while l <= r:
        if s[l] != s[r]:
            return False
        l += 1
        r -= 1
    return True
n = int(input())
s = str(n)
cnt = 0
while dx(str(n)) == False:
    n = n + dn(n)
    cnt += 1
print(cnt, n)
```

```python
#W4A9 
def kn(n):
    cnt = [0] * 11
    while n != 0: 
        r = n % 10
        if cnt[r] == 0:
            cnt[r] += 1
        else:
            return False
        n //= 10
    return True
if __name__ == '__main__':
    n = int(input())
    ck = False
    dem = 0
    for i in range(1, int(n**0.5)+1):
        if kn(i*i):
            dem += 1
            print(i*i, end = ' ')
    print('\n' + str(dem))
    
```

```python
#W4A10
cnt= [0] * (10**6)
def cz(n):
    cnt = 0
    while n != 1:
        if n % 2 == 0:
            n //= 2
            cnt += 1
        else:
            n = 3*n + 1
            cnt += 1
    return cnt
n = int(input())
for i in range(1, n+1):
    cnt[i] = cz(i)
ms = max(cnt)
for i in range(1, n+1):
    if cnt[i] == ms:
        print(i, end = ' ')
        break
print(ms)
```

```python
#w4A13
def tu(n):
    tong = 0
    for i in range(1, int(n**0.5)+1):
        if n % i == 0:
            tong += i
            if i != n // i:
                tong += n//i
    return tong - n 
if __name__ == '__main__':
    a, b = map(int, input().split())
    if tu(a) == b and tu(b) == a:
        print('true')
    else:
        print('false')
```

```python
#W4A14
def gcd(a, b):
    while b != 0:
        tmp = a % b
        a = b
        b = tmp
    return a
a, b = map(int, input().split())
print(gcd(a,b))
```
