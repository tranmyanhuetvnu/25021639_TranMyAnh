```python
#W7A1
def tknp(a, l, r,x):
    while l <= r:
        m = (l+r)//2
        if a[m] == x:
            return m
        elif a[m] < x:
            l = m + 1
        elif a[m] > x:
            r = m-1
    return -1
a = list(map(int, input().split()))
x = int(input())
print(tknp(a, 0, len(a)-1, x))
```

```python
#W7A2
def xh(a, x):
    return a.count(x)
a = list(map(int, input().split()))
x = int(input())
print(xh(a,x))
```
