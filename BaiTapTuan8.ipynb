```python
#W7A16
cnt = [0] * (10**6)
a = list(map(int, input().split()))
for x in a:
    cnt[x] += 1
res = 0
a = set(a)
for x in a:
    if cnt[x] < x:
        res += cnt[x]
    elif cnt[x] > x:
        res += cnt[x] - x
if res != 0:
    print('NO', res)
else:
    print('YES')
```

```python
#W7A18
a = list(map(int, input().split()))
sc =[]
sl = []
for x in a:
    if x % 2 == 0:
        sc.append(x)
    else:
        sl.append(x)
sc.sort()
sl.sort(reverse= True)
for x in sc:
    print(x, end = ' ')
for x in sl: 
    print(x, end = ' ')
```

```python
#W7A19
a = list(map(int, input().split()))
b = list(map(int, input().split()))
i = j = 0
r = []
if a == sorted(a) or b == sorted(b):
    while i < len(a) and j < len(b):
        if a[i] < b[j]:
            r.append(a[i])
            i+=1
        elif a[i] == b[j]:
            r.append(a[i])
            r.append(b[j])
            i+=1
            j+=1 
        else:
            r.append(b[j])
            j+=1
if a == sorted(a, reverse = True) or b == sorted(b, reverse = True):
    while i < len(a) and j < len(b):
        if a[i] > b[j]:
            r.append(a[i])
            i+=1
        elif a[i] == b[j]:
            r.append(a[i])
            r.append(b[j])
            i+=1
            j+=1 
        else:
            r.append(b[j])
            j+=1
while i < len(a):
    r.append(a[i])
    i+=1
while j < len(b):
    r.append(b[j])
    j+=1
print(r)
```

```python
#W7A22
n = int(input())
d = {}
for i in range(n):
    s = input()
    s = s.split()
    r = s[len(s)-1]
    d[r] = d.get(r, 0) + 1 
res = max(d, key=d.get)
print(res)
```
