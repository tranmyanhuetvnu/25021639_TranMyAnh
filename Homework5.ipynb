```python
#w6A16
n, m = map(int, input().split())
a = []
for i in range(n):
    b = list(map(int, input().split()))
    a.append(b)
for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:>4}", end = ' ')
    print()
```

```python
#w6A17 
n = int(input())
a = []
dcc = [] 
dcp = []
for i in range(n):
    b = list(map(int, input().split()))
    a.append(b)
for i in range(n):
    for j in range(n):
        if i == j:
            dcc.append(a[i][j])
        if n-1-i == j:
            dcp.append(a[i][j])
print(dcc, dcp)
```

```python
#w6A20
def tk(a, k):
    if k in a:
        return a.index(k)
    else:
        return -1
a = list(map(int, input().split()))
k = int(input())
print(tk(a, k))
```

```python
#W6A21
s = input().split()
d = {}
for i, w in enumerate(s):
    if w not in d:
        d[w] = []
    d[w].append(i) 
for x in d:
    d[x] = tuple(d[x])
print(d)
```

```python
#w6A22
def ln(a):
    b = max(a)
    return a.index(b)
a = list(map(int, input().split()))
print(ln(a))
```

```python
#W6A23
def xoa(a, x):
    b = []
    for i in range(len(a)):
        if a[i] >= x :
            b.append(a[i])
    return b
a = list(map(int, input().split())) 
x = int(input())
print(xoa(a, x))
```

```python
#W6A24
def d0(a):
    for i in range(len(a)):
        if a[i] == 0:
            a.pop(i)
            a.insert(0, 0)
    return a
a = list(map(int, input().split()))
print(d0(a))
```

```python
#W6A26
a = list(map(int, input().split()))
res = 0
for i in range(len(a)):
    if i != 0 and i != len(a) -1:
        if a[i] > a[i-1] and a[i+1] < a[i]:
            res += a[i]
print(res)
```

```python
#W6A28
def td(a): 
    n = len(a)   
    for i in range(1, n):
        if a[i] < a[i-1]:
            return False
    return True
def gd(a):  
    n = len(a)  
    for i in range(1, n):
        if a[i] > a[i-1]:
            return False
    return True
a = list(map(int, input().split()))
if gd(a) == True: 
    print('day giam thuc su')
if td(a) == True:
    print('day tang thuc su')
if td(a) == False and gd(a) == False:
    print('day vo danh')
```
