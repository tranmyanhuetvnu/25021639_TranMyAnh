```python
#w6A16
n, m = map(int, input().split())
a = []
for i in range(n):
    b = list(map(int, input().split()))
    a.append(b)
for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:>4}", end = ' ')
    print()
```

```python
#w6A17 
n = int(input())
a = []
dcc = [] 
dcp = []
for i in range(n):
    b = list(map(int, input().split()))
    a.append(b)
for i in range(n):
    for j in range(n):
        if i == j:
            dcc.append(a[i][j])
        if n-1-i == j:
            dcp.append(a[i][j])
print(dcc, dcp)
```

```python
#w6A18 
n, m, k = map(int, input().split())
a = []
for i in range(n):
    b = list(map(int, input().split()))
    a.append(b)
res= 0
for j in range(m):
    for i in range(n):
        if j == k:
            res += a[i][j]
print(res)
```

```python
#w6A19 
a = list(map(int, input().split()))
m = max(a)
for i in range(len(a)):
    if a[i] == m:
        print(i)
```
